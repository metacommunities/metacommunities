## Programming languages in use on Github

The programming languages are a primary component of life on Github. They organise important aspects of what is done there. They do not stand in isolation.


```python

import matplotlib.pyplot as plt, mpld3
import seaborn
from bs4 import BeautifulSoup
import pandas as pd
import re
mpld3.enable_notebook()
```

```python
repo_table = 'metacommunities:github_proper.repo_list'
```

```python
query = """SELECT repository_language, count(distinct(repository_url)) as language_count FROM [githubarchive:github.timeline] 
group by repository_language 
order by language_count desc;
"""

lang_df = pd.io.gbq.read_gbq(query)
```

```python
print(lang_df.repository_language.values)
lang_df[1:].head()
```

There are close to 200 'languages' in use, but most of them are not heavily used. The point here is to see how they structure platform-based work. 

Also, many repos have no language attributed to them: around 7.5M. But then language attributions are given for 13.75M.  So we are working with multiple language attributions? Or attributions that shift over time?
